```python
import numpy as np
```

```python
x=np.array([1.0,2.0,3.0]) # n차원 배열 생성은 np.array()

y=np.array([2.0,4.0,6.0]) # 배열간 산술연산 가능
print(x+y)
print(x-y)
print(x/y)
print(y*2.0) # 스칼라값이 행렬도 적당히 변환돼 계산이 배열의 원소의 수에 맞게 계산됨 -> 브로드캐스트!
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([10,20]) # [[10,20],[10,20]]으로 자동 변환돼
a*b # 형상이 다른 배열끼리도 적당히 계산이 가능함 -> 브로드캐스트!
```

```python
x = np.array([[51,55],[14,19],[0,4]])
x = x.flatten() # 1차원 배열로 평탄화
print(x)
```

```python
print(x[np.array([0,2,4])])
print(x[x>15])  
# 특정 조건을 만족하는 원소만 얻기
x>15
```

```python
import matplotlib.pyplot as plt

# 데이터 생성
x=np.arange(0,6,0.1) # 0-6까지 0.1 간격으로 생성
y = np.sin(x)

# 데이터 그래프화
plt.plot(x,y)
plt.show()
```

```python
# 데이터 준비
x = np.arange(0,6,0.1)
y1=np.sin(x)
y2=np.cos(x)

# 그래프화
plt.plot(x,y1, label="sin!") # label로 해당 plot에 라벨 붙임
plt.plot(x,y2, label="cos!")

plt.xlabel("x")
plt.ylabel("y")
plt.title("sin and cos we are family!")
plt.legend() # 차트 용례 표기
plt.show()
```
