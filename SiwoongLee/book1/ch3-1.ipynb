## 계단 함수
- 퍼셉트론에서 쓰이는 활성화 함수

```python
import numpy as np
```

```python
def step_function(x):
    y = x>0 # [true, false, false..] 양수인것만 true
    return y.astype(int) # astype은 bool -> int로
```

```python
x = np.array([-1.0,1.0,2.0])
step_function(x)
```

```python
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x>0, dtype=int)

x = np.arange(-5.0,5.0,0.1)
y=step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1) # y축 범위제한
plt.show()
```

## 시그모이드 함수
- S자형(sigmoid) 함수
- 퍼셉트론에선 계단 함수를 활성화 함수로 쓴다.
- 일반적인 신경망에서 가장 많이 쓰이는 활성화 함수가 시그모이드 함수이다.

```python
def sigmoid(x):
    return 1/ (1+np.exp(-x))








```

```python
x=np.array([-1.0,1.0,2.0])
sigmoid(x)
```

```python
x= np.arange(-5.0,5.0,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.show()
```


계단함수와의 공통점
1. 입력이 작으면 0에 가깝거나 같고 입력이 크다면 1에 가깝거나 같다.
2. 0~1 이내 => 뭐가좋냐? 결과를 '확률'로 사용해 분석가능하게 된다
3. 둘다 비선형 함수(= 함수 중 결과값이 입력의 상수배가 아닌 직선 1개로 그릴 수 없는 함수)이다.

신경망의 활성화함수는 비선형함수여야 한다. 
- 선형으로 쓰이면 결국 단층 레이어, 즉 퍼셉트론과 다를게 없어진다.
- 이렇게 되면 층을 여러게 쌓는, 즉 은닉층을 사용하는 이점이 없어짐.

## RELU 함수
- 입력이 0 이하면 0, 0을 넘는 입력값은 입력값 그대로를 출력.
- 요즘 많이 쓰이는 이유(기울기 소실 문제 해결 -> 시그모이드 양끝단 기울기가 0이 돼 학습이 잘 안됐음, but 얘는 기울기 1 고정)


```python
def relu(x):
    return np.maximum(0,x) # max()
```
